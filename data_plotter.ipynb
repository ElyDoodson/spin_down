```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from os import listdir

source = "d:/dev/spin_down/data/"
data_frames = [pd.read_csv(source + file_name, encoding="utf-8", delimiter="\t", comment = "#") for file_name in listdir(source) if "csv" in file_name]

data_frames[1].head()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (20,12))
for num, data in enumerate(data_frames):
    mass = data.M.tolist()
    period = data.Per.tolist()
    
    r,c = divmod(num,2)
    print(r,c)
    ax[r][c].invert_xaxis()

    ax[r][c].plot(mass, period, linestyle = "none", marker = "x")
    ax[r][c].set(xlim =(1.7,0.0), ylim = ([0,40.0]))

print(listdir(source))

+
```
